# Power Spectrum Density Analysis

```python
def power_in_freq_range(pwr_per_freq):
    freq_count = len(pwr_per_freq)
    power = 0
    for i in range(freq_count-1):
        #calculate area of trapizoid between frequency i and frequency i+1
        power = power + (pwr_per_freq[i] + pwr_per_freq[i+1]) / 2
    
    return power
```

```python
class mug:
    def __init__(self, handle_count, height, weight, color, fill):
        self.handle_count = handle_count
        self.height = height
        self.weight = weight
        self.color = color
        self.fill = fill
    
    def use_drink(self, sip_size):
        if self.fill != 'empty':
            self.fill = self.fill - sip_size
        if self.fill < 0:
            self.fill = 'empty'
    def check_fill(self):
        print(self.fill)
my_mug = mug(handle_count=1, height=10, weight=300, color='blue', fill=100)
my_mug.use_drink(110)
my_mug.check_fill()
```
